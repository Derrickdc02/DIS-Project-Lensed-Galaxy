# PQMass prior validation on Colab

Both prior draws and normalized PROBES arrays are read from Google Drive. Shared statistics are implemented by validate-pqmass.

```python
from google.colab import drive

drive.mount("/content/drive")
!git clone --depth 1 https://github.com/Derrickdc02/DIS-Project-Lensed-Galaxy.git /content/DIS-Project-Lensed-Galaxy
%cd /content/DIS-Project-Lensed-Galaxy
!pip install -e ".[validation]" -q
```

```python
import shlex
import subprocess
from pathlib import Path

DRIVE_ROOT = Path("/content/drive/MyDrive/DIS-Project-Lensed-Galaxy")
PRIOR = DRIVE_ROOT / "outputs" / "probes_final" / "prior_check"
DATA_DIR = DRIVE_ROOT / "data" / "gals_gband_norm"
RESULT = DRIVE_ROOT / "results" / "pqm_results.json"
RUN = False

command = [
    "validate-pqmass",
    "--prior", str(PRIOR),
    "--data-dir", str(DATA_DIR),
    "--output", str(RESULT),
    "--max-samples", "1000",
    "--num-refs", "100",
    "--re-tessellation", "200",
    "--pca-components", "50",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    RESULT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(command, check=True)
```

```python
import json

if RESULT.exists():
    print(json.dumps(json.loads(RESULT.read_text()), indent=2))
else:
    print(f"Result not found yet: {RESULT}")
```
