## Synthetic forward-model smoke test

The source is synthetic, so this cell is suitable for a CPU environment. It verifies that gradients can flow through the same SIE plus external-shear model used by posterior sampling.

```python
import torch

from lensing import build_lens_sim
from sample import lens_forward

size = 64
axis = torch.linspace(-1.0, 1.0, size)
yy, xx = torch.meshgrid(axis, axis, indexing="ij")
source = 2.0 * torch.exp(-((xx - 0.15) ** 2 + (yy + 0.10) ** 2) / 0.08) - 1.0
source.requires_grad_(True)

sim = build_lens_sim(image_size=size, pixelscale=0.04, source_pixelscale=0.028)
lensed = lens_forward(sim, source)
lensed.mean().backward()

print(f"source={tuple(source.shape)}, lensed={tuple(lensed.shape)}")
print(f"finite gradient={bool(torch.isfinite(source.grad).all())}")
```
