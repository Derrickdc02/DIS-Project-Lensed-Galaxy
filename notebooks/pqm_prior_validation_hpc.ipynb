# PQMass prior validation on Wilkes3

This notebook is a thin frontend to validate-pqmass. Generate prior draws first with scripts/run_sample_prior.sh.

```python
from pathlib import Path

cwd = Path.cwd().resolve()
REPO = next(path for path in (cwd, *cwd.parents) if (path / "pyproject.toml").is_file())
print(f"Repository: {REPO}")
```

```python
import shlex
import subprocess

PRIOR = REPO / "outputs" / "probes_final" / "prior_check"
DATA_DIR = REPO / "data" / "gals_gband_norm"
RESULT = REPO / "outputs" / "validation" / "pqm_results.json"
RUN = False

command = [
    "validate-pqmass",
    "--prior", str(PRIOR),
    "--data-dir", str(DATA_DIR),
    "--output", str(RESULT),
    "--max-samples", "1000",
    "--num-refs", "100",
    "--re-tessellation", "200",
    "--pca-components", "50",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    RESULT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(command, check=True)
```

```python
import json

if RESULT.exists():
    print(json.dumps(json.loads(RESULT.read_text()), indent=2))
else:
    print(f"Result not found yet: {RESULT}")
```

Interpretation: matched prior and data distributions should give chi2/dof near one and an approximately uniform p-value distribution. The real-versus-real sanity result must also remain compatible with a uniform p-value distribution.
