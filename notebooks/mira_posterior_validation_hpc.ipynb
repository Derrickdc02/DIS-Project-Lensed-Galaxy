# MIRA posterior calibration on Wilkes3

MIRA requires multiple truth sources. Point RUNS_ROOT at a directory containing one sampling run per source. Shared loading, PCA, baselines and scoring live in validate_mira.py.

```python
from pathlib import Path

cwd = Path.cwd().resolve()
REPO = next(path for path in (cwd, *cwd.parents) if (path / "pyproject.toml").is_file())
print(f"Repository: {REPO}")
```

```python
import shlex
import subprocess

RUNS_ROOT = REPO / "outputs" / "mira"
RESULT = REPO / "outputs" / "validation" / "mira_results.json"
RUN = False

command = [
    "validate-mira",
    "--runs-root", str(RUNS_ROOT),
    "--output", str(RESULT),
    "--pca-components", "50",
    "--num-runs", "200",
    "--num-bootstrap", "200",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    RESULT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(command, check=True)
```

```python
import json

if RESULT.exists():
    print(json.dumps(json.loads(RESULT.read_text()), indent=2))
else:
    print(f"Result not found yet: {RESULT}")
```

Use at least several independent truth sources. A single source is not an informative posterior-calibration test.
