# Posterior reconstruction on Colab

This notebook consumes a checkpoint and PROBES arrays stored on Google Drive. It calls the same sample-posterior CLI as Wilkes3.

```python
import shlex
import subprocess
from pathlib import Path

DRIVE_ROOT = Path("/content/drive/MyDrive/DIS-Project-Lensed-Galaxy")
DATA_DIR = DRIVE_ROOT / "data" / "gals_gband_norm"
OUTPUT_DIR = DRIVE_ROOT / "outputs" / "posterior_pick70"
CKPT = DRIVE_ROOT / "artifacts" / "latest.pt"
RUN = False

command = [
    "sample-posterior",
    "--data_dir", str(DATA_DIR),
    "--output_dir", str(OUTPUT_DIR),
    "--ckpt", str(CKPT),
    "--steps", "8000",
    "--n_post", "160",
    "--chunk", "32",
    "--pick", "70",
    "--noise_sigma", "0.02",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    subprocess.run(command, check=True)
```

```python
import torch

artifact = OUTPUT_DIR / "samples" / "posterior_draws.pt"
if artifact.exists():
    saved = torch.load(artifact, map_location="cpu", weights_only=False)
    posterior = saved["post"].float()
    print(f"draws={posterior.shape[0]}")
    print(f"mean posterior std={posterior.std(dim=0).mean():.4f}")
else:
    print(f"Run not found yet: {artifact}")
```
