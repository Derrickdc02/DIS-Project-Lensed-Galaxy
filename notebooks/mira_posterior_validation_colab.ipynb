```python
import shlex
import subprocess
from pathlib import Path

DRIVE_ROOT = Path("/content/drive/MyDrive/DIS-Project-Lensed-Galaxy")
RUNS_ROOT = DRIVE_ROOT / "outputs" / "mira"
RESULT = DRIVE_ROOT / "results" / "mira_results.json"
RUN = False

command = [
    "validate-mira",
    "--runs-root", str(RUNS_ROOT),
    "--output", str(RESULT),
    "--pca-components", "50",
    "--num-runs", "200",
    "--num-bootstrap", "200",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    RESULT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(command, check=True)
```

```python
import json

if RESULT.exists():
    print(json.dumps(json.loads(RESULT.read_text()), indent=2))
else:
    print(f"Result not found yet: {RESULT}")
```
