# Figure 2 out-of-distribution reproduction

Run the repository CLI against the public checkpoint on Drive. The command writes one deterministic figure and resumable intermediate chunks.

```python
from google.colab import drive

drive.mount("/content/drive")
!git clone --depth 1 https://github.com/Derrickdc02/DIS-Project-Lensed-Galaxy.git /content/DIS-Project-Lensed-Galaxy
%cd /content/DIS-Project-Lensed-Galaxy
!pip install -e ".[figures]" -q
```

```python
import shlex
import subprocess
from pathlib import Path

DRIVE_ROOT = Path("/content/drive/MyDrive/DIS-Project-Lensed-Galaxy")
CKPT = DRIVE_ROOT / "artifacts" / "latest.pt"
OUTPUT = DRIVE_ROOT / "results" / "figure2_ood.png"
RUN = False

command = [
    "reproduce-figure2",
    "--ckpt", str(CKPT),
    "--out", str(OUTPUT),
    "--steps", "8000",
    "--seed", "21",
]
print(shlex.join(command))
if RUN:
    OUTPUT.parent.mkdir(parents=True, exist_ok=True)
    subprocess.run(command, check=True)
```

```python
from IPython.display import Image, display

if OUTPUT.exists():
    display(Image(filename=str(OUTPUT)))
else:
    print(f"Figure not found yet: {OUTPUT}")
```
